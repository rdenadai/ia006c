#### **Exercício 1**

$~$

**Distribuição:**

|   X/Y   |  Y=0 |  Y=1  | Marg. X |
|:-------:|:----:|:-----:|:-------:|
|   X=0   |  1/6 |  3/8  |  13/24  |
|   X=1   |  1/8 |  1/3  |  11/24  |
| Marg. Y | 7/24 | 17/24 |    1    |

$~$

**a)** $P(X)$ e $P(Y)$

**Resposta:**

 - $P(X=x) = \{\frac{13}{24}, \frac{11}{24}\}$

 - $P(Y=y) = \{\frac{7}{24}, \frac{17}{24}\}$

$~$

**b)** $P(X=0|Y=0)$

$\frac{P({X=0}, {Y=0})}{P(Y=0)} = \frac{1}{6}\times\frac{24}{7} = \frac{24}{42} = \frac{4}{7}$

$\frac{P({X=1}, {Y=0})}{P(Y=0)} = \frac{1}{8}\times\frac{24}{7} = \frac{24}{56} = \frac{3}{7}$

**Resposta:**

 - $P(X=0|Y=0) = \frac{4}{7}$

$~$

**c)** $E[X]$ e $E[Y]$

$E[X] = \sum_kx_kP(x_k)$

$E[X] = 0 \times \frac{13}{24} + 1 \times \frac{11}{24}$

$E[Y] = 0 \times \frac{7}{24} + 1 \times \frac{17}{24}$


**Resposta:**

 - $E[X] = \frac{11}{24}$
 
 - $E[Y] = \frac{17}{24}$

$~$

**d)** São independentes? Por quê?

Resposta:

X e Y NÃO são independentes, pois a probabilidade do evento Y não afeta X, de acordo com a formulação:

$P(X|Y) = P(X)P(Y)$

Verificamos:

$P(X=x|Y=0) = P(X=x)P(Y=0)$

Por fim temos:

$P(X=0|Y=0) = P(X=0)P(Y=0) => \frac{4}{7} = \frac{13}{24}\times\frac{7}{24} => \frac{4}{7} \neq \frac{91}{576}$

$P(X=1|Y=0) = P(X=1)P(Y=0) => \frac{3}{7} = \frac{11}{24}\times\frac{7}{24} => \frac{3}{7} \neq \frac{77}{576}$

#### **Exercício 2**

$~$

**Distribuição:**

|   X/Y   | Y=0 | Y=1 | Marg. X |
|:-------:|:---:|:---:|:-------:|

$~$

**a)** $H(X), H(Y), H(X,Y)$

Sendo: $H(X) = -\Sigma_x p(x)log_2[p(x)]$

$H(X) = H(\frac{1}{4}, \frac{3}{4})$

$H(X) = -((\frac{1}{4} \times log_2[\frac{1}{4}]) + (\frac{3}{4} \times log_2[\frac{3}{4}]))$

$H(X) = -((\frac{1}{4} \times -2) + (\frac{3}{4} (log_2[3]-2)))$

$H(X) = -((-\frac{1}{2}) + (\frac{3}{4} (log_2[3]-2)))$

$H(X) = 0.8112$

---

$H(Y) = H(\frac{3}{8}, \frac{5}{8})$

$H(Y) = -((\frac{3}{8} \times log_2[\frac{3}{8}]) + (\frac{5}{8} \times log_2[\frac{5}{8}]))$

$H(Y) = -((\frac{3}{8} (log_2(3)-3)) + (\frac{5}{8} (log_2(5)-3)))$

$H(Y) = 0.9544$

---

Calculando $H(X, Y)$

Sendo: $H(X, Y) = -\Sigma_x\Sigma_y p(x, y) log_2[p(x, y)]$

$H(X, Y) = -((\frac{1}{4} log_2(\frac{1}{4}))+(\frac{3}{8} log_2(\frac{3}{8}))+(\frac{3}{8} log_2(\frac{3}{8})))$

$H(X, Y) = 1.5612$

---

$H(X, Y) = 1.5612$

$~$

**b)** $H(X|Y)$ e $H(Y|X)$

$H(Y|X) = -\Sigma_x\Sigma_y p(x, y) log_2[p(y|x)]$

$H(Y|X) = H(X, Y) - H(X)$

$H(X|Y) = H(X, Y) - H(Y)$

---

$P(Y=1 | X=0) = \frac{P(X=0, Y=1)}{P(X)} => \frac{1}{4}\times\frac{4}{1} = 1$

$P(Y=0 | X=1) = \frac{P(X=1, Y=0)}{P(X)} => \frac{3}{8}\times\frac{4}{3} = \frac{1}{2}$

$P(Y=1 | X=1) = \frac{P(X=1, Y=1)}{P(X)} = \frac{1}{2}$

$~$

$H(Y|X) = -( (\frac{1}{4} log_2(1)) + (\frac{3}{8} log_2(\frac{1}{2})) + (\frac{3}{8} log_2(\frac{1}{2})) )$

$H(Y|X) = -( (\frac{3}{8} \times -1) + (\frac{3}{8} \times -1) )$

$H(Y|X) = -( (-\frac{3}{8}) + (-\frac{3}{8}) )$

$H(Y|X) = 0.75$

$~$

$H(X|Y) = 1.5612 - 0.9544$

$H(X|Y) = 0.6068$

---

**Resposta:**

$H(Y|X) = 0.75$

$H(X|Y) = 0.6068$

Importação dos dados do *Australian Bureau of Meteorology* e sua apresentação.

Os dados, são definidos como uma série temporal onde em determinada data é apresentada a temperatura. Abaixo é apresentado os primeiros 10 registros dos 3650 itens.

```python
import csv
from datetime import datetime
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python

def convert(dt, tmp):
    date = datetime.strptime(dt, "%Y-%m-%d")
    return (
        datetime.strftime(date, "%d/%m/%Y"),
        float(tmp),
        datetime.timestamp(date),
        float(tmp),
        date.day,
        date.month,
        date.year,
    )

def import_data():
    data = None
    with open('daily-minimum-temperatures.csv') as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        data = [convert(*row) for i, row in enumerate(csv_reader) if i > 0]
    return np.array(data)

def kfold(data, k=1, shuffle=True):
    folds = []
    
    if shuffle:
        permutation = np.random.permutation(data.shape[0])
        data = data[permutation, :]
    
    mb = np.ceil(data.shape[0] / k).astype(np.int32)
    for r, _ in enumerate(range(k)):
        ini, end = r * mb, mb * (r+1)
        folds.append(data[ini:end, :])
    return np.array(folds)

# def linear_regression(x, y, lmbd=0):
#     T = x.T
#     inv = np.dot(T, x)
#     lmbd = np.eye(inv.shape[0], dtype=int) * lmbd
#     w = np.matmul(np.linalg.inv(inv + lmbd), np.matmul(T, y))
#     # yh = w0 + wk * x
#     w0 = w[0]
#     wk = np.concatenate((np.zeros((1,)), w[1:]))
#     return w, w0 + np.matmul(x, wk)


class LinearRegressionImpl:
    
    def __init__(self, x, y):
        self._x = np.copy(x)
        self._y = np.copy(y)
        self.w = np.random.uniform(size=self._x.shape)
        self.N = x.shape[0]
    
    def train(self, reg=0):
        T = self._x.T
        inv = np.dot(T, self._x)
        # inv = np.dot(x, T)
        reg = np.eye(inv.shape[0], dtype=int) * reg

    #     k = np.linalg.inv(inv + lmbd)
    #     a = np.matmul(T, k)
    #     w = np.matmul(a, y)
        self.w = np.matmul(np.linalg.inv(inv + reg), np.matmul(T, y))

        # Predict
        # yh = w0 + wk * x
        w0 = self.w[0]
        wk = np.concatenate((np.zeros((1, )), self.w[1:]))
        yh = w0 + np.matmul(self._x, wk)
        # Cost
        J0 = (1/self.N) * np.sum(np.square(y - yh))
        return self.w, yh, self.__MSE(self._y, yh)

    def __MSE(self, y, yh):
        return (1/self.N) * np.sum(np.square(y - yh))
    
    def predict(self, x, y):
        w0 = self.w[0]
        wk = np.concatenate((np.zeros((1, )), self.w[1:]))
        yh = w0 + np.matmul(x, wk)
        return yh, self.__MSE(y, yh)
```

```python
data = import_data()
df_data = np.copy(data)
data = data[:, 2:].astype(np.float32)
df = pd.DataFrame(df_data[:, 0:2], columns=["Data", "Temperature"])
df.head(10)
```
